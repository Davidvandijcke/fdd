```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(0, 1, 1000)

# Define the two functions
y1 = 0.5*(x-0.4)**3 + 0.1*x**2 + 0.35*x**(1/2) -  0.2*x**2 + 0.175
y2 = 0.4*(x-0.5)**3 + 0.5*x + 0.425

y_observed = y1 * (x < 0.5) + y2 * (x >= 0.5)
y_unobserved = y1 * (x >= 0.5) + y2 * (x < 0.5)


# Create the figure and the axes
fig, ax = plt.subplots(figsize=(7,4))

# Plot the two functions
ax.plot(x, y_observed, color="black",  linewidth = 2.5, label="Observed")
ax.plot(x, y_unobserved, color="black",  linestyle = "dashed", 
        alpha=0.7, label="Unobserved")


# Add the labels for the two functions
ax.text(0.1, 0.15, "E[Y(0)|X=x]", fontsize=12, va='center')
ax.text(0.01, 0.55, "E[Y(1)|X=x]", fontsize=12, va='center')

# grey dashed transparent vertical line at x=0.5
ax.axvline(x=0.5, color='grey', alpha=0.4, linestyle=(0,(5,10)))

# remove y and x axis ticks
ax.set_yticks([])
ax.set_xticks([])

# Set the labels of the axes
ax.set_xlabel('Running variable (X)', labelpad=10)
ax.set_ylabel('Outcome variable (Y)', labelpad=10)

plt.legend()
# Show the legend and set the x and y limits of the plot
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

# make plot prettier
sns.despine(right=True, top=True)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
plt.legend(frameon=False)

ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', labelsize=12)
ax.xaxis.set_tick_params(width=1.5)
ax.yaxis.set_tick_params(width=1.5)

# Show grid
ax.grid(False)


plt.tight_layout()



plt.savefig(os.path.join(figs_dir, 'lee_plot.pdf'), dpi=300, bbox_inches='tight')
```
